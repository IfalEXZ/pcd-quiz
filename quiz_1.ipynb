```python
from skimage import color
from skimage import io
import numpy as np

img = io.imread('https://akcdn.detik.net.id/community/media/visual/2023/10/18/avatar-the-last-airbender-3.png?w=700&q=90g')
imgGray = color.rgb2gray(img)

io.imshow(imgGray)
```

```python
height = img.shape[0]
width = img.shape[1]

print("Tinggi gambar:", height)
print("Lebar gambar:", width)
```

```python
# nomor3
img = np.array(img, dtype=np.uint8)

max_pixel = np.max(img)
min_pixel = np.min(img)

print("Nilai maksimum pixel:", max_pixel)
print("Nilai minimum pixel:", min_pixel)
```

```python
# nomor4
# Normalisasi gambar
normalized_image = img / 255.0

io.imshow(normalized_image)
```

```python
flip_horizontal = np.flip(img, axis=1)
flip_vertical = np.flip(img, axis=0)

io.imshow(flip_horizontal)
io.show()

io.imshow(flip_vertical)
io.show()
```

```python
# Menambahkan noise Gaussian
mean = 0
stddev = 0.1
gaussian_noise = np.random.normal(mean, stddev, img.shape)
noisy_image_gaussian = np.clip(img + gaussian_noise, 0, 1)

# Menambahkan noise Salt and Pepper
salt_pepper_prob = 0.05
salt_pepper_noise = np.random.choice([0, 1, 2], size=img.shape, p=[1 - salt_pepper_prob, salt_pepper_prob / 2, salt_pepper_prob / 2])
noisy_image_salt_pepper = np.copy(img)
noisy_image_salt_pepper[salt_pepper_noise == 1] = 0
noisy_image_salt_pepper[salt_pepper_noise == 2] = 1


io.imshow(noisy_image_gaussian)
io.show()

io.imshow(noisy_image_salt_pepper)
io.show()
```

```python
# Random Crop
random_crop_size = (250, 250)
height, width, _ = img.shape
x = np.random.randint(0, width - random_crop_size[1] + 1)
y = np.random.randint(0, height - random_crop_size[0] + 1)
random_crop = img[y:y+random_crop_size[0], x:x+random_crop_size[1], :]

# Center Crop
center_crop_size = (250, 250)
start_x = (width - center_crop_size[1]) // 2
start_y = (height - center_crop_size[0]) // 2
end_x = start_x + center_crop_size[1]
end_y = start_y + center_crop_size[0]
center_crop = img[start_y:end_y, start_x:end_x, :]

# Menampilkan gambar hasil crop
io.imshow(random_crop)
io.show()

io.imshow(center_crop)
io.show()
```
